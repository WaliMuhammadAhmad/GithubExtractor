```python
repos = [
    'Train4llama',
    'Test4llama',
    'Eval4llama',
]

dir = 'Llama2'
```

```python
for repo in repos:
  repo = Repository(local_dir=repo,clone_from=f'CodexAI/{repo}', repo_type='dataset')
  repo.git_pull()
```

```python
files = os.listdir('Train4llama/data')

for f in files:
  f = os.path.join('Train4llama/data',f)
  if f.endswith(".parquet"):
    shutil.move(f,'Eval4llama/data')
```
