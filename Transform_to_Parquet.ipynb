```python
for dir in os.listdir(repo_name):
    dir_path = os.path.join(repo_name, dir)

    if os.path.isdir(dir_path) and dir != '.git' and dir != '.gitattributes':
        shutil.rmtree(dir_path)

print("Folders removed, keeping only .parquet, .git, and .gitattributes")
```
