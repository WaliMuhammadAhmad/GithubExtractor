```python
import os
import shutil
from datasets import load_dataset
from huggingface_hub import login, Repository
from access_token import validate_token
```

```python
login(validate_token(),add_to_git_credential=True)
```

```python
repo_url="CodexAI/dataset"

repo_name="dataset"
```

```python
repo = Repository(local_dir=repo_name,clone_from=repo_url, repo_type=repo_name)
repo.git_pull()
```

```python
dataset = load_dataset(repo_name)
```

```python
for dir in os.listdir(repo_name):
    dir_path = os.path.join(repo_name, dir)

    if os.path.isdir(dir_path) and dir != '.git' and dir != '.gitattributes':
        shutil.rmtree(dir_path)

print("Folders removed, keeping only .parquet, .git, and .gitattributes")
```
